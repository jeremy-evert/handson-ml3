# Forecast Uncertainty Analysis

In this notebook we explore **forecast uncertainty**.

Weather forecasts represent predictions about future atmospheric
conditions. Because the atmosphere is a complex chaotic system,
forecast accuracy decreases as the prediction horizon increases.

This notebook investigates the relationship between:

- forecast horizon
- prediction accuracy

We will compare forecast predictions against **actual observations**
recorded by airport weather stations (METAR data).

Our goal is to measure:

• forecast error  
• how error changes over time  
• how uncertainty grows with prediction horizon

## Learning Objectives

By the end of this notebook students should understand:

1. Forecasts are predictions, not measurements
2. Prediction error increases with time horizon
3. Real-world data contains noise and uncertainty
4. Forecast systems have measurable reliability limits
