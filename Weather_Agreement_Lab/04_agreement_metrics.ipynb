# 03 – Agreement Analysis Between Agencies

## Objective

This notebook quantifies how much different observational agencies agree on hourly temperature during 2012.

We will compute:

- Pearson correlation
- Root Mean Squared Error (RMSE)
- Mean bias
- Seasonal breakdown of error
- Diurnal cycle agreement

We will visualize:

- Time series overlays
- Difference plots
- Scatter plots with regression lines
- Heatmaps of correlation

This notebook answers the core scientific question:

"How consistent are independent observational networks measuring the same physical phenomenon?"
